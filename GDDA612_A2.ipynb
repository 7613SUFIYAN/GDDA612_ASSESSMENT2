```python
import pandas as pd
import pymongo

# Load data from CSV into a Pandas DataFrame
df = pd.read_csv('tidy_data.csv')

# Connect to MongoDB
client = pymongo.MongoClient('mongodb+srv://vhorasufiyan76:Sufiyan%4011_99241@cluster0.4paa608.mongodb.net/')
db = client['Churn_db']  
collection = db['churn_data']
# Convert DataFrame to dictionary for easier insertion into MongoDB
data = df.to_dict(orient='records')
print("MongoDB Connection Successful.")

# Insert data into MongoDB collection
collection.insert_many(data)
print("Dataset imported into MongoDB collection successfully.")

# Count documents in the collection
count_documents = collection.count_documents({})
print(f"Total documents in collection: {count_documents}")

# Find and print the first document
first_document = collection.find_one()
print("\nFirst document in collection:\n", first_document)

# Sort documents by 'age' in ascending order
sorted_docs = collection.find().sort('Age', pymongo.ASCENDING)

# Print the first few sorted documents
print("\nSorted documents by 'Age':")
for doc in sorted_docs.limit(5):
    print(doc)

# Count documents in the collection
count_documents = collection.count_documents({})
print(f"Total documents in collection: {count_documents}")
```

```python
# Aggregation Pipeline
pipeline = [
    {
        "$bucket": {
            "groupBy": "$Age",
            "boundaries": [20, 30, 40, 50, 60, 70, 80, 90, 100],  
            "default": "Other",
            "output": {
                "average_estimated_salary": {"$avg": "$EstimatedSalary"}
            }
        }
    },
    {
        "$sort": {"_id": 1}
    }
]
# Run aggregation
result = collection.aggregate(pipeline)

# Print the result
print("Average EstimatedSalary by Age Group:")
for doc in result:
    
    print(doc)
```

```python
# Define the filter and update
filter = {"Geography": "Spain"}  
update = {"$inc": {"CreditScore": 100}}

# Execute the update operation
result = collection.update_many(filter, update)

# Print the result
print(f"Matched count: {result.matched_count}, Modified count: {result.modified_count}")
```
